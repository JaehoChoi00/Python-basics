Write a function getBASIC() which takes no arguments, and does the following: it should keep reading lines from input using a while loop; when it reaches the end it should return the whole program in the form of a list of strings. (Hints: about lists and stopping)

```python
def getBASIC():
    program = []
    while True:
        line = input()
        program.append(line)
        if (line.split()[-1] == "END"):
            return program
```

Once we have read the program, we need to be able to move from line to line in the program. To accomplish this, we ask you to write the following subroutine.

Define a function findLine(prog, target) to perform the following. Assume prog is a list of strings containing a BASIC program, like the type generated by getBASIC(); assume target is a string containing a line number, which is the target of a GOTO statement. The function should return the index i (a number between 0 and len(prog)-1) such that prog[i] is the line whose label equals target.

Sample input/output: If you call

findLine(['10 GOTO 20','20 END'], '10')

then the output should be 0, since item 0 of the list is the line with label 10.

```python
def findLine(prog, target):
    for line in range (len(prog)):
        if(prog[line].split()[0] == target):
            return line

print(findLine(['10 GOTO 20','20 END'], '10'))
print(findLine(['4 GOTO 12', '12 GOTO 99', '22 GOTO 22', '99 GOTO 12', '200 END'], '22'))
```

Sub-task 3: Smart Simulation
In the previous two exercises, we handled the input routine and the searching task. These will be useful to make writing the main program shorter. However, there is still a major question: how can we actually solve the original problem? The most straightforward way would be to simulate the BASIC program:

    let prog be the BASIC program (an array of strings)
    start a counter called location at 0, since we begin on the first line of the program
    while True,

        if prog[location] is the END line, return "success" and stop.
        let T be the target string indicated in prog[location]
        let the new value of location be findLine(prog, T)

But, there is a major problem: this doesn't detect infinite loops, and if the BASIC program has an infinite loop, then the Python program will also loop forever. What we wanted was that the program should print "infinite loop" in this situation. We leave this problem for you to solve; we give a hint below.

Write a function execute(prog) to do the following. Assume prog is a list of strings containing the BASIC program, like before. Then, your function should return either "success" or "infinite loop" depending on whether the program terminates, or loops forever. Important: you should assume the procedure findLine(prog, target) defined in sub-task 2 is already defined, you do not need to re-write it.

```python
def execute(prog):
    location = 0
    while True:
        if location==len(prog)-1: return "success"
        T = prog[location].split()[-1]
        prevLocation = location
        location = findLine(prog, T)
        if (prevLocation > location): return "infinite loop"

print(execute(['10 GOTO 21', '21 GOTO 37', '37 GOTO 21', '40 END']))
```

```python
def execute(prog):
    location = 0
    visited = []
    while True:
        if location==len(prog)-1: return "success"
        if location in visited: return "infinite loop"
        visited.append(location)
        T = prog[location].split()[-1]
        location = findLine(prog, T)

print(execute(['10 GOTO 21', '21 GOTO 37', '37 GOTO 21', '40 END']))
```

```python
import mpmath

def calculate_pi_chudnovsky_steps(steps):
    # Dynamically scale precision: ~14 digits per step + a buffer of 10 digits
    mpmath.mp.dps = (steps * 14) + 10
    
    # This will hold our running total for the sum (the Sigma symbol)
    sigma_sum = mpmath.mpf(0)
    
    # Loop from k = 0 up to (steps - 1)
    for k in range(steps):
        # Numerator parts
        num_sign = (-1)**k
        num_fact = mpmath.factorial(6 * k)
        num_linear = 545140134 * k + 13591409
        numerator = num_sign * num_fact * num_linear
        
        # Denominator parts
        den_fact1 = mpmath.factorial(3 * k)
        den_fact2 = mpmath.factorial(k)**3
        den_power = mpmath.mpf(640320)**(3 * k + 1.5)
        denominator = den_fact1 * den_fact2 * den_power
        
        # Add the current term to our running sum
        sigma_sum += numerator / denominator
    
    # Multiply the final sum by 12 to get 1/Pi
    one_over_pi = 12 * sigma_sum
    
    # Invert it to get Pi
    pi_estimate = 1 / one_over_pi
    return pi_estimate

# --- TEST THE CODE ---
# Change this number to run more steps!
num_steps = 1000

print(f"Pi calculation using {num_steps} steps:")
print(calculate_pi_chudnovsky_steps(num_steps))

print("\nActual value of Pi (matched to the same precision):")
print(mpmath.mp.pi)
```

```python
import mpmath
import sys
import time
import gc

sys.setrecursionlimit(100000)
sys.set_int_max_str_digits(50000000)

def chudnovsky_bs(a, b):
    if b - a == 1:
        if a == 0:
            return 1, 1, 13591409
        else:
            P = (6 * a - 5) * (2 * a - 1) * (6 * a - 1)
            Q = a**3 * 26680512000000
            T = P * (545140134 * a + 13591409)
            if a % 2 == 1:
                T = -T
        return P, Q, T
    else:
        mid = (a + b) // 2
        P1, Q1, T1 = chudnovsky_bs(a, mid)
        P2, Q2, T2 = chudnovsky_bs(mid, b)
        return P1 * P2, Q1 * Q2, Q2 * T1 + P1 * T2

def stress_test_pi(steps):
    expected_digits = steps * 14
    print(f"--- INIT STRESS TEST: Target {expected_digits:,} Digits ---")
    
    # Allocate decimal precision
    mpmath.mp.dps = expected_digits + 10
    
    # Profile Phase 1: Integer Tree Multiplication
    print("Phase 1: Building and reducing Binary Splitting Tree...")
    t0 = time.perf_counter()
    P, Q, T = chudnovsky_bs(0, steps)
    t1 = time.perf_counter()
    print(f"  -> Tree reduction finished in {t1 - t0:.4f} seconds.")
    
    # Profile Phase 2: Arbitrary Precision Decimal Division
    print("Phase 2: Executing high-precision floating division...")
    t2 = time.perf_counter()
    top = mpmath.mpf(Q) * 426880
    bottom = mpmath.mpf(T)
    pi = (top * mpmath.sqrt(10005)) / bottom
    t3 = time.perf_counter()
    print(f"  -> Division & square root finished in {t3 - t2:.4f} seconds.")
    
    # Profile Phase 3: Text Conversion (Memory Wall)
    print("Phase 3: Serializing floating data to raw string format...")
    t4 = time.perf_counter()
    pi_str = str(pi)
    t5 = time.perf_counter()
    print(f"  -> String generation finished in {t5 - t4:.4f} seconds.")
    
    # Free up RAM memory actively
    del P, Q, T, top, bottom, pi
    gc.collect()
    
    # Save Out
    print("Phase 4: Streaming string bytes straight to disk...")
    with open("extremedigits.txt", "w") as f:
        f.write(pi_str)
    
    total_time = (t5 - t0) + (t3 - t2)
    print(f"--- COMPLETE: Total Time Taken: {total_time:.2f} seconds ---")
    print(f"Result file 'extremedigits.txt' successfully saved to workspace.")

# CHANGE THIS VALUE TO HIT THE WALL
# 20,000 steps = ~280,000 digits
# 75,000 steps = ~1,050,000 digits (The 1-Million-Digit Milestone)
num_steps = 20000 
stress_test_pi(num_steps)
```
