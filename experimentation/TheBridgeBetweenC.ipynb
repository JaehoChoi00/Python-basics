```python
import sys
import time

print("Downloading package...")
for i in range(0, 101, 5):
    print(chr(13) + "Progress: [", end = '')

    for j in range (0, i//5):
        print('#', end = '')
    
    for j in range (i//5, 20):
        print('.', end = '')
    
    print(f"] {i}%", end = '')
    sys.stdout.flush()
    time.sleep(0.15)

print("\nDownloadComplete!")
```

We can simplify the process more by using the print format. And how python allows arithmetics within the print function.

```python
import sys
import time

print("Downloading package...")

for i in range (0, 101, 5):
    print(f"{chr(13)}Progress: [{'#' * (i//5)}{'.' * (20 - i//5)}] {i}%", end = '')
    sys.stdout.flush()
    time.sleep(0.15)

print("\nDownload Complete!")
```

A fun challenge is to add a rotating spinner using these:

/ | - \

```python
print(ord('|'))
print(ord('/'))
print(ord('-'))
print(ord('\\'))
spinner = ['|', '/', '-', '\\']
spinnerASCII = [124, 47, 45, 92]

for i in range(0, 101, 5):
    print(f"{chr(13)}{spinner[(i//5)%4]}=<o>={chr(spinnerASCII[(i//5)%4])} ", end = '')
    sys.stdout.flush()
    time.sleep(0.05)
```

```python
import sys
import time

spinner = ['|', '/', '-', '\\']
print("Downloading package...")

for i in range (0, 101, 5):
    print(f"{chr(13)}Progress: [{'#' * (i//5)}{'.' * (20 - i//5)}] {i}% {spinner[(i//5)%4]}", end = '')
    sys.stdout.flush()
    time.sleep(0.15)

print("\nDownload Complete!")
```
